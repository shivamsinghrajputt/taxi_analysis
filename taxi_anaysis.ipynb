```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
data=data.filter(['pickup_datetime','passenger_count','pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','trip_duration'])
```

```python
data
```

```python
#extract date, month, day_name, hour

data['month']=pd.to_datetime(data['pickup_datetime']).dt.month_name()
data['date']=pd.to_datetime(data['pickup_datetime']).dt.day
data['week_day']=pd.to_datetime(data['pickup_datetime']).dt.day_name()
data['pickup_hours']=pd.to_datetime(data['pickup_datetime']).dt.hour
```

```python
data
```

```python
#analysis in which date maximun and least use of taxis.

sns.countplot(data=data, x='date')
plt.show()
```

```python
#weak_day analysis...

sns.countplot(data=data, x='week_day')
plt.show()

#friday and saturday used maximum nunmbers of taxis....
```

```python
#analysis on months.....

sns.countplot(data=data, x='month')
plt.show()

#in march used maximun no. of taxis...
```

```python
#nalysis over time, in which time taxis use maximum....
sns.countplot(data=data,x='pickup_hours')
plt.show()

print('At which time more taxis are used:\n',data['pickup_hours'].mode())

#from 4-5 time taxi use least and 18-19 time  used most.
```
